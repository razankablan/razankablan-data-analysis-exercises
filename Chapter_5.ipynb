```python
import pandas as pd
import matplotlib.pyplot as plt

if 'date' in fb_stock.columns:
    fb_stock['date'] = pd.to_datetime(fb_stock['date'])
    fb_stock.set_index('date', inplace=True)

# Calculate the prior day's closing price using shift()
fb_stock['prev_close'] = fb_stock['close'].shift(1)

# Calculate daily differences
fb_stock['after_hours_change'] = fb_stock['open'] - fb_stock['prev_close']
fb_stock['intraday_change'] = fb_stock['close'] - fb_stock['open']

# Create two subplots side by side
plt.figure(figsize=(14, 5))

# Left subplot: after-hours change ---
plt.subplot(1, 2, 1)
plt.plot(fb_stock.index, fb_stock['after_hours_change'], color='purple')
plt.title("After-Hours Change (Open - Prior Close)")
plt.xlabel("Date")
plt.ylabel("Price Difference ($)")

# Right subplot: intraday change ---
plt.subplot(1, 2, 2)
plt.plot(fb_stock.index, fb_stock['intraday_change'], color='green')
plt.title("Intraday Change (Close - Open)")
plt.xlabel("Date")
plt.ylabel("Price Difference ($)")

# Adjust layout and show
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

if 'date' in fb_stock.columns:
    fb_stock['date'] = pd.to_datetime(fb_stock['date'])
    fb_stock.set_index('date', inplace=True)

# Compute prior day's close
fb_stock['prev_close'] = fb_stock['close'].shift(1)

# After-hours daily change
fb_stock['after_hours_change'] = fb_stock['open'] - fb_stock['prev_close']

# Aggregate monthly net effect
monthly_after_hours = fb_stock['after_hours_change'].resample('M').sum()  # sum of daily after-hours moves

# Create side-by-side plots
plt.figure(figsize=(14,5))

# Left subplot: daily after-hours change (line plot) ---
plt.subplot(1,2,1)
plt.plot(fb_stock.index, fb_stock['after_hours_change'], color='purple')
plt.title("Daily After-Hours Change (Open - Prior Close)")
plt.xlabel("Date")
plt.ylabel("Price Difference ($)")

# Right subplot: monthly net effect (bar plot) ---
plt.subplot(1,2,2)
monthly_after_hours.plot(kind='bar', color='orange', width=0.8)
plt.title("Monthly Net After-Hours Effect")
plt.xlabel("Month")
plt.ylabel("Net Price Change ($)")
plt.xticks(rotation=45)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

if 'date' in fb_stock.columns:
    fb_stock['date'] = pd.to_datetime(fb_stock['date'])
    fb_stock.set_index('date', inplace=True)

# Compute prior day's close
fb_stock['prev_close'] = fb_stock['close'].shift(1)

# Daily after-hours change
fb_stock['after_hours_change'] = fb_stock['open'] - fb_stock['prev_close']

# Monthly net effect
monthly_after_hours = fb_stock['after_hours_change'].resample('M').sum()

# Bar colors: green for gains, red for drops
colors = ['green' if x >= 0 else 'red' for x in monthly_after_hours]

# Side-by-side plots
plt.figure(figsize=(14,5))

# Left subplot: daily after-hours change
plt.subplot(1,2,1)
plt.plot(fb_stock.index, fb_stock['after_hours_change'], color='purple')
plt.title("Daily After-Hours Change (Open - Prior Close)")
plt.xlabel("Date")
plt.ylabel("Price Difference ($)")

# Right subplot: monthly net effect
plt.subplot(1,2,2)
monthly_after_hours.plot(kind='bar', color=colors, width=0.8)
plt.title("Monthly Net After-Hours Effect")
plt.xlabel("Month")
plt.ylabel("Net Price Change ($)")
plt.xticks(rotation=45)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

if 'date' in fb_stock.columns:
    fb_stock['date'] = pd.to_datetime(fb_stock['date'])
    fb_stock.set_index('date', inplace=True)

# Compute prior day's close
fb_stock['prev_close'] = fb_stock['close'].shift(1)

# Daily after-hours change
fb_stock['after_hours_change'] = fb_stock['open'] - fb_stock['prev_close']

# Monthly net effect
monthly_after_hours = fb_stock['after_hours_change'].resample('M').sum()

# Bar colors: green for gains, red for drops
colors = ['green' if x >= 0 else 'red' for x in monthly_after_hours]

# Three-letter month abbreviations for x-axis
month_labels = monthly_after_hours.index.strftime('%b')

# Side-by-side plots
plt.figure(figsize=(14,5))

# Left subplot: daily after-hours change (line plot)
plt.subplot(1,2,1)
plt.plot(fb_stock.index, fb_stock['after_hours_change'], color='purple')
plt.title("Daily After-Hours Change (Open - Prior Close)")
plt.xlabel("Date")
plt.ylabel("Price Difference ($)")

# Right subplot: monthly net effect (colored bars)
plt.subplot(1,2,2)
plt.bar(month_labels, monthly_after_hours, color=colors, width=0.6)
plt.title("Monthly Net After-Hours Effect")
plt.xlabel("Month")
plt.ylabel("Net Price Change ($)")

# Adjust layout
plt.tight_layout()
plt.show()
```
